# Random Forest

```python
from sklearn.model_selection import GridSearchCV

grid = {'n_estimators' : [200,400,600],
        'max_depth' :[7,8,9],
        'min_samples_leaf' : [15,30,70]}

rand_model = RandomForestClassifier(criterion='gini')
grid_search = GridSearchCV(estimator = rand_model , param_grid = grid ,cv = 5, verbose = True)
grid_search.fit(x_train , y_train)
```

```python
grid_search.best_estimator_
```

```python
df_model = RandomForestClassifier(criterion='gini' , max_depth = 9 , min_samples_leaf = 15 , n_estimators = 600)
rf = df_model.fit(x_train , y_train)
```

```python
train = rf.predict_proba(x_train)
test = rf.predict_proba(x_test)
```

```python
print("Performance of train :",round(roc_auc_score(y_train , train[: , 1]) , 4))
print("Performance of test :",round(roc_auc_score(y_test , test[: , 1]) , 4))
```

```python
t = [300 , 500 , 700 ]
ml = [15 , 30 ,70]
md = [7,8,9]
model_performance_data = []

for i in t:
  for j in ml:
    for k in md:

        mp = RandomForestClassifier(criterion='gini' , max_depth = k , min_samples_leaf = j, n_estimators = i)
        dt_model = mp.fit(x_train , y_train)
        test_pred = dt_model.predict_proba(x_test)
        train_pred = dt_model.predict_proba(x_train)

        t1 = "NO. of t-"+str(i)+" max_depth-"+str(k)+" min_leaf-"+str(j)
        t2 = [t1 , round(roc_auc_score(y_train , train_pred[: , 1]),4) , round(roc_auc_score(y_test , test_pred[: , 1]),4)]

        model_performance_data.append(t2)
```

```python
model = pd.DataFrame(model_performance_data)
```

```python
model.rename(columns = {0: "Values" , 1 : "Trains_AUC", 2: "Test_AUC"} , inplace = True)
```

```python
model.sort_values(by = ["Test_AUC", "Trains_AUC"] , ascending = [False,False]).head(10)
```

# Logistic Regression

```python
import statsmodels.api as sm
x_train['constant'] = 1
x_test['constant'] = 1
log_reg = sm.Logit(y_train, x_train).fit()
```

```python
log_reg.summary()
```

```python
x_train.head(2)
x_train['pred_prob'] = log_reg.predict(x_train)
x_test['pred_prob'] = log_reg.predict(x_test)
x_train.head(3)
```

```python
from sklearn.metrics import roc_curve, auc
# Compute ROC curve and area the curve for logistic
fpr1, tpr1, thresholds1 = roc_curve(y_train, x_train['pred_prob'])
roc_auc1 = auc(fpr1, tpr1)
gini1 = (2*roc_auc1)-1
print("Area under the ROC curve for train dataset : %f" % roc_auc1)
print("Gini Co-efficent for train dataset : %f" % gini1 )
print("--"*15)
```

```python
# Compute ROC curve and area the curve for logistic
fpr2, tpr2, thresholds2 = roc_curve(y_test, x_test['pred_prob'])
roc_auc2 = auc(fpr2, tpr2)
gini2 = (2*roc_auc2)-1
print("Area under the ROC curve for test dataset : %f" % roc_auc2)
print("Gini Co-efficent for train dataset : %f" % gini2 )
```

```python
import pylab as pl
# Plot ROC curve
pl.clf()
pl.plot(fpr1, tpr1, label='ROC curve for training data (area = %0.2f)'
% roc_auc1)
pl.plot(fpr2, tpr2, label='ROC curve for test data (area = %0.2f)' %
roc_auc2)
pl.plot([0, 1], [0, 1], 'k--')
pl.xlim([0.0, 1.0])
pl.ylim([0.0, 1.0])
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.title('Receiverrating characteristic example')
pl.legend(loc="lower right")
pl.show()
```

```python
print("Confusion Matrix")
cm = metrics.confusion_matrix(y_test , diab_test_predict , labels = [1,0])

df_cm = pd.DataFrame(cm , index = [i for i in ["1","0"]],
                     columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm , annot = True)
```
